```python
def load_coordinate_data(trainID, valID, testID):
    idxTrain_5f, xTrain_5f, yTrain_5f = [], [], []
    idxVal_5f, xVal_5f, yVal_5f = [], [], []
    idxTest_5f, xTest_5f, yTest_5f = [], [], []
    
    for fold_idx in range(n_FOLD):
        idxTrain, xTrain, yTrain = [], [], []
        idxVal, xVal, yVal = [], [], []
        idxTest, xTest, yTest = [], [], []

        train_index = trainID[fold_idx]
        val_index = valID[fold_idx]
        test_index = testID[fold_idx]

        for pid in train_index: ### ===================================== Train =====================================
            csv_path = KeypointsROOT + pid + '/Sit-and-Stand_20240314/' + pid + '_7_sit2stand_timeSeries_resampling.csv'
            if not os.path.exists(csv_path):
                print(f'{pid} time series data CSV does not exit.')
            else:
                df = pd.read_csv(csv_path, header=None)
                n_feature, n_timestamp = df.shape[0], df.shape[1]
                data = df.values
                xData = data.reshape(n_feature, n_timestamp).tolist()
                yLabel = Label_dict[int(pid)]
                idxTrain.append(int(pid))
                xTrain.append(xData)
                yTrain.append(yLabel)
        
        for pid in val_index: ### ===================================== Validation =====================================
            csv_path = KeypointsROOT + pid + '/Sit-and-Stand_20240314/' + pid + '_7_sit2stand_timeSeries_resampling.csv'
            if not os.path.exists(csv_path):
                print(f'{pid} time series data CSV does not exit.')
            else:
                df = pd.read_csv(csv_path, header=None)
                n_feature, n_timestamp = df.shape[0], df.shape[1]
                data = df.values
                xData = data.reshape(n_feature, n_timestamp).tolist()
                yLabel = Label_dict[int(pid)]
                idxVal.append(int(pid))
                xVal.append(xData)
                yVal.append(yLabel)
        
        for pid in test_index: ### ===================================== Test =====================================
            csv_path = KeypointsROOT + pid + '/Sit-and-Stand_20240314/' + pid + '_7_sit2stand_timeSeries_resampling.csv'
            if not os.path.exists(csv_path):
                print(f'{pid} time series data CSV does not exit.')
            else:
                df = pd.read_csv(csv_path, header=None)
                n_feature, n_timestamp = df.shape[0], df.shape[1]
                data = df.values
                xData = data.reshape(n_feature, n_timestamp).tolist()
                yLabel = Label_dict[int(pid)]
                idxTest.append(int(pid))
                xTest.append(xData)
                yTest.append(yLabel)

        idxTrain_5f.append(np.array(idxTrain))
        xTrain_5f.append(np.array(xTrain))
        yTrain_5f.append(np.array(yTrain))
        idxVal_5f.append(np.array(idxVal))
        xVal_5f.append(np.array(xVal))
        yVal_5f.append(np.array(yVal))
        idxTest_5f.append(np.array(idxTest))
        xTest_5f.append(np.array(xTest))
        yTest_5f.append(np.array(yTest))
    return idxTrain_5f, xTrain_5f, yTrain_5f, idxVal_5f, xVal_5f, yVal_5f, idxTest_5f, xTest_5f, yTest_5f


idxTrain_5f_Sit2Stand, xTrain_5f_Sit2Stand, yTrain_5f_Sit2Stand, \
    idxVal_5f_Sit2Stand, xVal_5f_Sit2Stand, yVal_5f_Sit2Stand, \
        idxTest_5f_Sit2Stand, xTest_5f_Sit2Stand, yTest_5f_Sit2Stand = load_coordinate_data(trainID_5f, validateID_5f, testID_5f)
```

## Training

```python
# n_PERIOD = 6
n_PERIOD = 25

n_MODEL = 3

# MAX_EPOCHS = 15
# MAX_EPOCHS = 25
# MAX_EPOCHS = 30
MAX_EPOCHS = 20
# MAX_EPOCHS = 50
# BATCH_SIZE = 16
BATCH_SIZE = 4
Threshold = 0.5

best_threshold_5f_sit2stand_List = []
best_f1score_5f_sit2stand_List = []
```

```python
def build_model_structure():
    model = Sequential()
    model.add(Conv2D(2, kernel_size=(1, 4), activation='relu', strides=1, padding='SAME', input_shape=(int(n_FEATURE), int(n_PERIOD), 1), name='first_conv')) ### grayscale image
    # model.add(MaxPooling2D(pool_size=(1, 2)))
    model.add(Conv2D(2, kernel_size=(1, 4), activation='relu', strides=1, name='last_conv')) 
    # model.add(MaxPooling2D(pool_size=(1, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(8, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))
    return model
```

### Train & Save models

```python
best_threshold_5f_List_Sit2Stand = []
best_f1score_5f_List_Sit2Stand = []

def show_training_loss_and_accuracy(history):
    plt.figure(figsize=(10, 3))
    plt.subplot(121)
    plt.title('Training loss')
    plt.xlabel('Epoch')
    plt.ylabel('loss')
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='val')
    plt.legend()
    
    plt.subplot(122)
    plt.title('Training accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('accuracy')
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='val')
    plt.legend()
    plt.show()


def find_best_threshold_tuning(yTrue, yPred_prob):
    scores = []
    thresholds = np.arange(0.40, 0.60, 0.01)
    for t in thresholds:
        yPred = (yPred_prob >= t).astype('int')
        score = f1_score(yTrue, yPred)
        scores.append(score)
    
    ix = np.argmax(scores)
    best_threshold = thresholds[ix]
    best_f1score = scores[ix]
    return best_threshold, best_f1score


def training_model_Validation_data_Save_model():
    idxTrain_5f, xTrain_5f, yTrain_5f = idxTrain_5f_Sit2Stand, xTrain_5f_Sit2Stand, yTrain_5f_Sit2Stand
    idxVal_5f, xVal_5f, yVal_5f = idxVal_5f_Sit2Stand, xVal_5f_Sit2Stand, yVal_5f_Sit2Stand
    
    for fold_idx in range(n_FOLD):
        idxTrain_1f, xTrain_1f, yTrain_1f = idxTrain_5f[fold_idx], xTrain_5f[fold_idx], yTrain_5f[fold_idx]
        idxVal_1f, xVal_1f, yVal_1f = idxVal_5f[fold_idx], xVal_5f[fold_idx], yVal_5f[fold_idx]

        xTrain_1f = xTrain_1f.reshape(-1, int(n_FEATURE), int(n_PERIOD), 1) ### shape = (n_samples, 13, 6, 1)
        xVal_1f = xVal_1f.reshape(-1, int(n_FEATURE), int(n_PERIOD), 1)

        best_auc = 0
        best_model_idx = 0
        for model_idx in range(n_MODEL):
            print(f'[Sit-to-Stand] Training for [Fold {fold_idx+1}] model {model_idx+1} ...')
            model = build_model_structure()
            model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
            history = model.fit(xTrain_1f, yTrain_1f, 
                                epochs=MAX_EPOCHS,
                                batch_size=BATCH_SIZE,
                                validation_data=(xVal_1f, yVal_1f), 
                                shuffle=True, verbose=0)
            yPred_val = model.predict(xVal_1f)
            aucScore_val = roc_auc_score(yVal_1f, yPred_val)
            if aucScore_val > best_auc:
                best_auc = aucScore_val
                best_model = model
                # show_training_loss_and_accuracy(history)

        ### Get threshold with Validation Data
        yPred_val = best_model.predict(xVal_1f)
        best_threshold, best_f1score = find_best_threshold_tuning(yVal_1f, yPred_val)
        best_threshold = round(best_threshold, 4)

        ### Confusion Matrix for Validation Data
        yPred_val = [x>best_threshold for x in yPred_val]
        print(f'{confusion_matrix(yVal_1f, yPred_val)}\n')

        best_f1score_5f_List_Sit2Stand.append(best_f1score)
        best_threshold_5f_List_Sit2Stand.append(best_threshold)
            
        ### Save best model -----------------------------------------------------------------------------
        if not os.path.exists(SAVEMODEL_FOLDER):
            os.makedirs(SAVEMODEL_FOLDER)
        best_model.save(f'{SAVEMODEL_FOLDER}sit2stand_model_Fold{fold_idx+1}.h5')
        
        ### Show model structure ------------------------------------------------------------------------
        if fold_idx == n_FOLD-1:
            print(best_model.summary())
        
    ### Save Threshold & F1-Score ------------------------------------------------------------------------
    if not os.path.exists(SAVEPARAMETER_FOLDER):
        os.makedirs(SAVEPARAMETER_FOLDER)
    threshold_file_name = SAVEPARAMETER_FOLDER + 'sit2stand_threshold_5fold.csv'
    f1score_file_name = SAVEPARAMETER_FOLDER + 'sit2stand_f1Score_5fold.csv'
    with open(threshold_file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(best_threshold_5f_List_Sit2Stand)
    with open(f1score_file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(best_f1score_5f_List_Sit2Stand)


training_model_Validation_data_Save_model()
```

```python
actual_label_List_Sit2Stand= []
predict_label_List_Sit2Stand = []
result_dict_Sit2Stand = {}

def check_performance_with_best_threshold(idxList, answerList, predictionList, best_threshold):
    uniqueID_test, countsPair_test = np.unique(idxList, return_counts=True)
    n_Correct = 0
    
    ### Loop through each pid in test list
    for i in range(len(uniqueID_test)):
        pid = uniqueID_test[i]
        nPairs = countsPair_test[i]
        prob = 0
        
        ### get the actual label (target) of this pid
        actual_label = Label_dict[pid]

        for j in range(len(idxList)):
            tmp_pid = idxList[j]
            if tmp_pid == pid:
                prob += round(predictionList[j][0], 4)
        
        ### calculate the average prediction prob of this pid   
        avegProb = prob / nPairs
        
        if avegProb >= best_threshold:
            predict_label = 1
            actual_label_List_Sit2Stand.append(actual_label)
            predict_label_List_Sit2Stand.append(predict_label)
        else:
            predict_label = 0
            actual_label_List_Sit2Stand.append(actual_label)
            predict_label_List_Sit2Stand.append(predict_label)
        
        if predict_label == actual_label:
            n_Correct += 1
        
        result_dict_Sit2Stand[pid] = {}
        result_dict_Sit2Stand[pid]["avegProb"] = round(avegProb, 4)
        result_dict_Sit2Stand[pid]["n_pairs"] = nPairs
        result_dict_Sit2Stand[pid]["prediction"] = predict_label
        result_dict_Sit2Stand[pid]["actual"] = actual_label
    return n_Correct, len(uniqueID_test)
```

```python
from mlxtend.plotting import plot_confusion_matrix

def show_confusion_matrix(mode, actual_label_List, predict_label_List):
    actual_label_array = np.array(actual_label_List)
    predict_label_array = np.array(predict_label_List)

    ### F1-Score
    print(f'==============  Classification Report  ==============')
    print(classification_report(actual_label_array, predict_label_array))

    ### Confusion Matrix
    matrix = confusion_matrix(actual_label_array, predict_label_array)
    TN, FP, FN, TP = matrix.ravel()

    value_accuracy = round((TP+TN)/(TP+TN+FP+FN), 4)
    value_sensitivity = round(TP/(TP+FN), 4)
    value_specificity = round(TN/(TN+FP), 4)
    value_gmean = round(np.sqrt(value_sensitivity*value_specificity), 4)
    value_precision = round(TP/(TP+FP), 4)
    value_f1_score = round(2*value_precision*value_sensitivity/(value_precision+value_sensitivity), 2)
    value_f2_measure = round(5*value_precision*value_sensitivity / (4*value_precision+value_sensitivity), 2)

    print(f'==================  Performance  ====================')
    print(f'0) TN: {TN}, FP: {FP}, FN: {FN}, TP: {TP}')
    print(f'1) Accuracy: {round(value_accuracy*100, 2)}%')
    print(f'2) Sensitivity (Recall): {round(value_sensitivity*100, 2)}%')
    print(f'3) Specificity: {round(value_specificity*100, 2)}%')
    print(f'4) G-Mean: {round(value_gmean*100, 2)}%')
    print(f'5) Precision: {round(value_precision*100, 2)}%')
    print(f'6) F1-Score: {value_f1_score}')
    print(f'7) F2-Measure: {value_f2_measure}\n')
    
    classList = ['NC', 'AD']
    fig, ax = plot_confusion_matrix(conf_mat=matrix, figsize=(3, 3), colorbar=True, class_names=classList, fontcolor_threshold=0.7)
    plt.xlabel('Predictions', fontsize=10)
    plt.ylabel('Actuals', fontsize=10)
    plt.title(f'{mode}', fontsize=12)
    plt.show()
```

```python
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = Model( model.inputs, [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()


def show_GradCAM_result():
    for fold_idx in range(n_FOLD):
        model = load_model(f'{SAVEMODEL_FOLDER}sit2stand_model_Fold{fold_idx+1}.h5')

        idxTest_1f, xTest_1f, yTest_1f = idxTest_5f_Sit2Stand[fold_idx], xTest_5f_Sit2Stand[fold_idx], yTest_5f_Sit2Stand[fold_idx]

        for test_idx in range(len(idxTest_1f)):
            cur_pid = idxTest_1f[test_idx]

            # if cur_pid==60 or cur_pid==36:
            # if cur_pid==16:
            if cur_pid==18:
                test_data = xTest_1f[test_idx]
                width, height = test_data.shape[1], test_data.shape[0]
                test_data = np.expand_dims(test_data, axis=0) ### shape=(1, 12, 50)

                test_prob = model.predict(test_data.reshape(-1, int(n_FEATURE), width, 1)) ### grayscale image
                # test_prob = model.predict(test_data.reshape(-1, int(n_FEATURE), int(n_PERIOD), 1)) ### grayscale image
                # test_prob = model.predict(test_data.reshape(-1, int(n_FEATURE), int(n_PERIOD), 3)) ### RGB image
                
                
                heatmap = make_gradcam_heatmap(test_data, model, 'last_conv')
                heatmap = np.uint8(255*heatmap)
                heatmap = cv2.resize(heatmap, (width, height)) ### shape=(13, 25)

                plt.imshow(heatmap, cmap='viridis', aspect='auto', interpolation='nearest', vmin=heatmap.min(), vmax=heatmap.max(), alpha=1.0)

                feature_name_list = ['angle',
                                     'xHipAdd', 'xKneeAdd', 'xAnkleAdd', 'xHipMinus', 'xKneeMinus', 'xAnkleMinus', 
                                     'yHipAdd', 'yKneeAdd', 'yAnkleAdd', 'yHipMinus', 'yKneeMinus', 'yAnkleMinus']
                print(f'------------------------------- [Fold {fold_idx+1}] [pid {cur_pid}] Prob: {test_prob[0][0]} -------------------------------')
                print(f'                     {result_dict_Sit2Stand[int(cur_pid)]}')
                
                plt.figure(figsize=(25, 12))
                for featureIdx in range(n_FEATURE):
                    plt.subplot(5, 3, featureIdx + 1)
                    plt.title(f'Feature: {feature_name_list[featureIdx]}', fontsize=10)

                    plt.imshow(np.expand_dims(heatmap[featureIdx], axis=0), cmap='viridis', aspect='auto', interpolation='nearest', 
                               vmin=heatmap.min(), vmax=heatmap.max(), alpha=1.0, 
                               extent=[0, width, round(test_data[0][featureIdx,:].min()*1.05), round(test_data[0][featureIdx,:].max()*1.05)])
                    plt.plot(test_data[0][featureIdx, :], 'k')
                plt.show()  


show_GradCAM_result()  
```

```python
def build_model_structure():
    model = Sequential()
    model.add(Conv2D(2, kernel_size=(1, 4), activation='relu', strides=1, padding='SAME', input_shape=(int(n_FEATURE), int(n_PERIOD), 1), name='first_conv')) ### grayscale image
    # model.add(MaxPooling2D(pool_size=(1, 2)))
    model.add(Conv2D(2, kernel_size=(1, 4), activation='relu', strides=1, name='last_conv')) 
    # model.add(MaxPooling2D(pool_size=(1, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(8, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))
    return model
```

```python
best_threshold_5f_List_Stand2Sit = []
best_f1score_5f_List_Stand2Sit = []

def show_training_loss_and_accuracy(history):
    plt.figure(figsize=(10, 3))
    plt.subplot(121)
    plt.title('Training loss')
    plt.xlabel('Epoch')
    plt.ylabel('loss')
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='val')
    plt.legend()
    
    plt.subplot(122)
    plt.title('Training accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('accuracy')
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='val')
    plt.legend()
    plt.show()

def find_best_threshold_tuning(yTrue, yPred_prob):
    scores = []
    thresholds = np.arange(0.40, 0.60, 0.01)
    for t in thresholds:
        yPred = (yPred_prob >= t).astype('int')
        score = f1_score(yTrue, yPred)
        scores.append(score)
    
    ix = np.argmax(scores)
    best_threshold = thresholds[ix]
    best_f1score = scores[ix]
    return best_threshold, best_f1score


def training_model_Validation_data_Save_model():
    idxTrain_5f, xTrain_5f, yTrain_5f = idxTrain_5f_Stand2Sit, xTrain_5f_Stand2Sit, yTrain_5f_Stand2Sit
    idxVal_5f, xVal_5f, yVal_5f = idxVal_5f_Stand2Sit, xVal_5f_Stand2Sit, yVal_5f_Stand2Sit
    
    for fold_idx in range(n_FOLD):
        idxTrain_1f, xTrain_1f, yTrain_1f = idxTrain_5f[fold_idx], xTrain_5f[fold_idx], yTrain_5f[fold_idx]
        idxVal_1f, xVal_1f, yVal_1f = idxVal_5f[fold_idx], xVal_5f[fold_idx], yVal_5f[fold_idx]

        xTrain_1f = xTrain_1f.reshape(-1, int(n_FEATURE), int(n_PERIOD), 1) ### shape = (n_samples, 13, 7, 1)
        xVal_1f = xVal_1f.reshape(-1, int(n_FEATURE), int(n_PERIOD), 1)

        best_auc = 0
        best_model_idx = 0
        for model_idx in range(n_MODEL):
            print(f'[Stand-to-Sit] Training for [Fold {fold_idx+1}] model {model_idx+1} ...')
            model = build_model_structure()
            model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
            history = model.fit(xTrain_1f, yTrain_1f, 
                                epochs=MAX_EPOCHS,
                                batch_size=BATCH_SIZE,
                                validation_data=(xVal_1f, yVal_1f), 
                                shuffle=True, verbose=0)
            yPred_val = model.predict(xVal_1f)
            aucScore_val = roc_auc_score(yVal_1f, yPred_val)
            if aucScore_val > best_auc:
                best_auc = aucScore_val
                best_model = model
                # show_training_loss_and_accuracy(history)

        ### Get threshold with Validation Data
        yPred_val = best_model.predict(xVal_1f)
        best_threshold, best_f1score = find_best_threshold_tuning(yVal_1f, yPred_val)
        best_threshold = round(best_threshold, 4)

        best_f1score_5f_List_Stand2Sit.append(best_f1score)
        best_threshold_5f_List_Stand2Sit.append(best_threshold)
            
        ### Save best model -----------------------------------------------------------------------------
        if not os.path.exists(SAVEMODEL_FOLDER):
            os.makedirs(SAVEMODEL_FOLDER)
        best_model.save(f'{SAVEMODEL_FOLDER}stand2sit_model_Fold{fold_idx+1}.h5')

        ### Show model structure ------------------------------------------------------------------------
        if fold_idx == n_FOLD-1:
            print(best_model.summary())
        
    ### Save Threshold & F1-Score ------------------------------------------------------------------------
    if not os.path.exists(SAVEPARAMETER_FOLDER):
        os.makedirs(SAVEPARAMETER_FOLDER)
    threshold_file_name = SAVEPARAMETER_FOLDER + 'stand2sit_threshold_5fold.csv'
    f1score_file_name = SAVEPARAMETER_FOLDER + 'stand2sit_f1Score_5fold.csv'
    with open(threshold_file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(best_threshold_5f_List_Stand2Sit)
    with open(f1score_file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(best_f1score_5f_List_Stand2Sit)
```

```python
def check_performance_with_best_threshold(idxList, answerList, predictionList, best_threshold, mode):
    uniqueID_test, countsPair_test = np.unique(idxList, return_counts=True)
    n_Correct = 0
    
    ### Loop through each pid in test list
    for i in range(len(uniqueID_test)):
        pid = uniqueID_test[i]
        nPairs = countsPair_test[i]
        prob = 0
        
        ### get the actual label (target) of this pid
        actual_label = Label_dict[pid]

        for j in range(len(idxList)):
            tmp_pid = idxList[j]
            if tmp_pid == pid:
                prob += round(predictionList[j][0], 4)
        
        ### calculate the average prediction prob of this pid   
        avegProb = prob / nPairs
        
        if avegProb >= best_threshold:
            predict_label = 1
            actual_label_List_Stand2Sit.append(actual_label)
            predict_label_List_Stand2Sit.append(predict_label)
        else:
            predict_label = 0
            actual_label_List_Stand2Sit.append(actual_label)
            predict_label_List_Stand2Sit.append(predict_label)
    
        if predict_label == actual_label:
            n_Correct += 1
        
        result_dict_Stand2Sit[pid] = {}
        result_dict_Stand2Sit[pid]["avegProb"] = round(avegProb, 4)
        result_dict_Stand2Sit[pid]["n_pairs"] = nPairs
        result_dict_Stand2Sit[pid]["prediction"] = predict_label
        result_dict_Stand2Sit[pid]["actual"] = actual_label
    return n_Correct, len(uniqueID_test)
```

```python
actual_label_List_Stand2Sit= []
predict_label_List_Stand2Sit = []

result_dict_Stand2Sit = {}

def load_model_and_evaluate_model():
    metrics_auc_5f = []
    metrics_acc_5f = []
    metrics_loss_5f = []
    nb_correction_5f = []
    nb_testing_5f = []
    
    for fold_idx in range(n_FOLD):
        ### Load trained model ---------------------------------------------------
        model = load_model(f'{SAVEMODEL_FOLDER}stand2sit_model_Fold{fold_idx+1}.h5')

        ### Load Test data -------------------------------------------------------
        idxTest_1f, xTest_1f, yTest_1f = idxTest_5f_Stand2Sit[fold_idx], xTest_5f_Stand2Sit[fold_idx], yTest_5f_Stand2Sit[fold_idx]
        xTest_1f = xTest_1f.reshape(-1, int(n_FEATURE), int(n_PERIOD), 1)

        ### Evaluate model with Test data -------------------------------------------------------
        yPred = model.predict(xTest_1f)
        auc = roc_auc_score(yTest_1f, yPred)
        metrics_auc_5f.append(round(auc, 2))
        scores = model.evaluate(xTest_1f, yTest_1f, verbose=0)
        metrics_acc_5f.append(round(scores[1], 2))
        metrics_loss_5f.append(round(scores[0], 2))

        # ### Check model performance with Threshold(=0.5) or best_threshold
        threshold = best_threshold_5f_List_Stand2Sit[fold_idx]
        nb_correction, nb_testing = check_performance_with_best_threshold(idxTest_1f, yTest_1f, yPred, threshold, "Stand2Sit")
        nb_correction_5f.append(nb_correction)
        nb_testing_5f.append(nb_testing)
    return nb_correction_5f, nb_testing_5f, metrics_auc_5f, metrics_acc_5f, metrics_loss_5f


correctionList_stand2sit, testNumList_stand2sit, aucList_stand2sit, accList_stand2sit, lossList_stand2sit = load_model_and_evaluate_model()
```

```python
from mlxtend.plotting import plot_confusion_matrix

def show_confusion_matrix(mode, actual_label_List, predict_label_List):
    actual_label_array = np.array(actual_label_List)
    predict_label_array = np.array(predict_label_List)

    ### F1-Score
    print(f'==============  Classification Report  ==============')
    print(classification_report(actual_label_array, predict_label_array))

    ### Confusion Matrix
    matrix = confusion_matrix(actual_label_array, predict_label_array)
    TN, FP, FN, TP = matrix.ravel()

    value_accuracy = round((TP+TN)/(TP+TN+FP+FN), 4)
    value_sensitivity = round(TP/(TP+FN), 4)
    value_specificity = round(TN/(TN+FP), 4)
    value_gmean = round(np.sqrt(value_sensitivity*value_specificity), 4)
    value_precision = round(TP/(TP+FP), 4)
    value_f1_score = round(2*value_precision*value_sensitivity/(value_precision+value_sensitivity), 2)
    value_f2_measure = round(5*value_precision*value_sensitivity / (4*value_precision+value_sensitivity), 2)

    print(f'==================  Performance  ====================')
    print(f'0) TN: {TN}, FP: {FP}, FN: {FN}, TP: {TP}')
    print(f'1) Accuracy: {round(value_accuracy*100, 2)}%')
    print(f'2) Sensitivity (Recall): {round(value_sensitivity*100, 2)}%')
    print(f'3) Specificity: {round(value_specificity*100, 2)}%')
    print(f'4) G-Mean: {round(value_gmean*100, 2)}%')
    print(f'5) Precision: {round(value_precision*100, 2)}%')
    print(f'6) F1-Score: {value_f1_score}')
    print(f'7) F2-Measure: {value_f2_measure}\n')
    
    classList = ['NC', 'AD']
    fig, ax = plot_confusion_matrix(conf_mat=matrix, figsize=(3, 3), colorbar=True, class_names=classList, fontcolor_threshold=0.7)
    plt.xlabel('Predictions', fontsize=10)
    plt.ylabel('Actuals', fontsize=10)
    plt.title(f'{mode}', fontsize=12)
    plt.show()
```

## Grad-CAM

```python
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = Model( model.inputs, [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()


def show_GradCAM_result():
    for fold_idx in range(n_FOLD):
        model = load_model(f'{SAVEMODEL_FOLDER}stand2sit_model_Fold{fold_idx+1}.h5')

        idxTest_1f, xTest_1f, yTest_1f = idxTest_5f_Stand2Sit[fold_idx], xTest_5f_Stand2Sit[fold_idx], yTest_5f_Stand2Sit[fold_idx]

        for test_idx in range(len(idxTest_1f)):
            cur_pid = idxTest_1f[test_idx]

            # if cur_pid==60 or cur_pid==36:
            # if cur_pid==16:
            if cur_pid==18:
                test_data = xTest_1f[test_idx]
                width, height = test_data.shape[1], test_data.shape[0]
                test_data = np.expand_dims(test_data, axis=0) ### shape=(1, 12, 50)

                test_prob = model.predict(test_data.reshape(-1, int(n_FEATURE), width, 1)) ### grayscale image
                # test_prob = model.predict(test_data.reshape(-1, int(n_FEATURE), int(n_PERIOD), 1)) ### grayscale image
                # test_prob = model.predict(test_data.reshape(-1, int(n_FEATURE), int(n_PERIOD), 3)) ### RGB image
                
                
                heatmap = make_gradcam_heatmap(test_data, model, 'last_conv')
                heatmap = np.uint8(255*heatmap)
                heatmap = cv2.resize(heatmap, (width, height)) ### shape=(13, 25)

                plt.imshow(heatmap, cmap='viridis', aspect='auto', interpolation='nearest', vmin=heatmap.min(), vmax=heatmap.max(), alpha=1.0)

                feature_name_list = ['angle',
                                     'xHipAdd', 'xKneeAdd', 'xAnkleAdd', 'xHipMinus', 'xKneeMinus', 'xAnkleMinus', 
                                     'yHipAdd', 'yKneeAdd', 'yAnkleAdd', 'yHipMinus', 'yKneeMinus', 'yAnkleMinus']
                print(f'------------------------------- [Fold {fold_idx+1}] [pid {cur_pid}] Prob: {test_prob[0][0]} -------------------------------')
                print(f'                     {result_dict_Stand2Sit[int(cur_pid)]}')
                
                plt.figure(figsize=(25, 12))
                for featureIdx in range(n_FEATURE):
                    plt.subplot(5, 3, featureIdx + 1)
                    plt.title(f'Feature: {feature_name_list[featureIdx]}', fontsize=10)

                    plt.imshow(np.expand_dims(heatmap[featureIdx], axis=0), cmap='viridis', aspect='auto', interpolation='nearest', 
                               vmin=heatmap.min(), vmax=heatmap.max(), alpha=1.0, 
                               extent=[0, width, round(test_data[0][featureIdx,:].min()*1.05), round(test_data[0][featureIdx,:].max()*1.05)])
                    plt.plot(test_data[0][featureIdx, :], 'k')
                plt.show()  


show_GradCAM_result()  
```

```python
actual_label_List_combined = []
predict_label_List_combined = []

result_dict_combined = {}
result_dict_combined_sorted = {}


def combine_two_predict_results_with_new_threshold():    
    if len(result_dict_Sit2Stand_sorted)==len(result_dict_Stand2Sit_sorted): print(f'[Info] Same number of subjects.')
    else: print(f'[Info] Something wrong with the number of subjects.')

    n_Correct = 0
    n_Subject = len(result_dict_Sit2Stand_sorted)
    
    for i in range(n_FOLD):
        testID_1f = testID_5f[i]
        new_threshold = best_threshold_5f_combined_List[i]
        
        for j in range(len(testID_1f)):
            pid = int(testID_1f[j])
            
            prob1 = result_dict_Sit2Stand_sorted[pid]["avegProb"]
            prob2 = result_dict_Stand2Sit_sorted[pid]["avegProb"]
            
            ### Method 1 ======================================== update 20240516
            # new_prob = round((prob1+prob2)/2, 4)
            
            ### Method 2 ========================================
            f1 = best_f1score_5f_sit2stand_List[i]
            f2 = best_f1score_5f_stand2sit_List[i]
            if f1==0 and f2==0:
                f1 = 1
                f2 = 1
            new_prob = np.average([prob1, prob2], weights=[f1, f2])
            
            label1 = result_dict_Sit2Stand_sorted[pid]["prediction"]
            label2 = result_dict_Stand2Sit_sorted[pid]["prediction"]
            
            actual_label = result_dict_Sit2Stand_sorted[pid]["actual"]
            actual_label_List_combined.append(actual_label)
            
            if new_prob >= new_threshold:
                predict_label = 1
                predict_label_List_combined.append(predict_label)
                if predict_label == actual_label:
                    n_Correct += 1
            else:
                predict_label = 0
                predict_label_List_combined.append(predict_label)
                if predict_label == actual_label:
                    n_Correct += 1
                    
            result_dict_combined[pid] = {}
            result_dict_combined[pid]["newProb"] = new_prob
            result_dict_combined[pid]["newLabel"] = predict_label
            result_dict_combined[pid]["label1"] = label1
            result_dict_combined[pid]["label2"] = label2
            result_dict_combined[pid]["actual"] = actual_label
    return n_Correct, n_Subject


nb_correction_combined, nb_testing_combined = combine_two_predict_results_with_new_threshold()
```

```python
combined_metrics_5fold_dict = {}

def load_model_and_evaluate_five_folds_metrics():
    metrics_5fold_dict = {}
    
    for fold_idx in range(n_FOLD):
        testID_1f = testID_5f[fold_idx]
        yTest_1f = yTest_5f_Stand2Sit[fold_idx]
        
        predict_label_List = []
        actual_label_List = []
        yPred = []
        for i, pid in enumerate(testID_1f):
            pid = int(pid)
            label = Label_dict[pid]
            threshold = best_threshold_5f_combined_List[fold_idx]
            f1 = best_f1score_5f_sit2stand_List[fold_idx]
            f2 = best_f1score_5f_stand2sit_List[fold_idx]
            if f1==0 and f2==0:
                f1 = 1
                f2 = 1
            
            prob1 = result_dict_Sit2Stand_sorted[pid]["avegProb"]
            prob2 = result_dict_Stand2Sit_sorted[pid]["avegProb"]
            new_prob = np.average([prob1, prob2], weights=[f1, f2])

            if new_prob >= threshold: predict_label = 1
            else: predict_label = 0
            
            predict_label_List.append(predict_label)
            actual_label_List.append(label)
            
            yPred.append(new_prob)
            
            
        predict_label_array = np.array(predict_label_List)
        actual_label_array = np.array(actual_label_List)
        matrix = confusion_matrix(actual_label_array, predict_label_array)
        TN, FP, FN, TP = matrix.ravel()
        value_accuracy = round((TP+TN)/(TP+TN+FP+FN), 4)
        value_sensitivity = round(TP/(TP+FN), 4)
        value_specificity = round(TN/(TN+FP), 4)
        value_precision = round(TP/(TP+FP), 4)
        value_f1_score = round(2*value_precision*value_sensitivity/(value_precision+value_sensitivity), 2)
        if np.isfinite(value_f1_score)==False: value_f1_score = 0
        auc = roc_auc_score(yTest_1f, yPred)
        
        classList = ['Non-AD', 'AD']
        matrix = confusion_matrix(actual_label_array, predict_label_array)
        TN, FP, FN, TP = matrix.ravel()
        fig, ax = plot_confusion_matrix(conf_mat=matrix, figsize=(2, 2), colorbar=True, class_names=classList, fontcolor_threshold=0.7)
        plt.xlabel('Predictions', fontsize=10)
        plt.ylabel('Actuals', fontsize=10)
        plt.title(f'Fold {fold_idx+1}', fontsize=12)
        plt.show()
        
        metrics_5fold_dict[fold_idx] = {}
        metrics_5fold_dict[fold_idx]['AUC'] = round(auc, 4)
        metrics_5fold_dict[fold_idx]['Accuracy'] = value_accuracy
        metrics_5fold_dict[fold_idx]['Sensitivity'] = value_sensitivity
        metrics_5fold_dict[fold_idx]['Specificity'] = value_specificity
        metrics_5fold_dict[fold_idx]['Precision'] = value_precision
        metrics_5fold_dict[fold_idx]['F1-Score'] = value_f1_score
    return metrics_5fold_dict


combined_metrics_5fold_dict = load_model_and_evaluate_five_folds_metrics()
```
