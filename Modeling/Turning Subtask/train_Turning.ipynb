```python
turning_dict_normalize = {}

def generate_scaled_data_for_training(trainID_5f, valID_5f, testID_5f, flag):
    X_train_5f, X_val_5f, X_test_5f = [], [], []
    
    for i in range(n_FOLD):
        trainID_1f = trainID_5f[i]
        valID_1f = valID_5f[i]
        testID_1f = testID_5f[i]
        
        tmp_train_list = []
        tmp_val_list = []
        tmp_test_list = []
        for pid in trainID_1f: ### type(pid) = string
            pid_int = int(pid)
            label = Label_dict[pid_int]
            p1 = int(turning_dict[pid]['period1'])
            p2 = int(turning_dict[pid]['period2'])
            if flag=='scaled':
                p1 = turning_dict_normalize[pid]['period1']
                p2 = turning_dict_normalize[pid]['period2']
            tmp_train_list.append([pid, p1, p2, label])
        
        for pid in valID_1f: ### type(pid) = string
            pid_int = int(pid)
            label = Label_dict[pid_int]
            p1 = int(turning_dict[pid]['period1'])
            p2 = int(turning_dict[pid]['period2'])
            if flag=='scaled':
                p1 = turning_dict_normalize[pid]['period1']
                p2 = turning_dict_normalize[pid]['period2']
            tmp_val_list.append([pid, p1, p2, label])   

        for pid in testID_1f:
            pid_int = int(pid)
            label = Label_dict[pid_int]
            p1 = int(turning_dict[pid]['period1'])
            p2 = int(turning_dict[pid]['period2'])
            if flag=='scaled':
                p1 = turning_dict_normalize[pid]['period1']
                p2 = turning_dict_normalize[pid]['period2']
            tmp_test_list.append([pid, p1, p2, label])
            
        df_train_1f = pd.DataFrame(tmp_train_list, columns=['pid', 'period1', 'period2', 'label'])
        df_val_1f = pd.DataFrame(tmp_val_list, columns=['pid', 'period1', 'period2', 'label'])
        df_test_1f = pd.DataFrame(tmp_test_list, columns=['pid', 'period1', 'period2', 'label'])
        
        X_train_5f.append(df_train_1f)
        X_val_5f.append(df_val_1f)
        X_test_5f.append(df_test_1f)    
    return X_train_5f, X_val_5f, X_test_5f


df_train_5f, df_val_5f, df_test_5f = generate_scaled_data_for_training(trainID_5f, validateID_5f, testID_5f, "origin")
# df_train_5f_scaled, df_val_5f_scaled, df_test_5f_scaled = generate_scaled_data_for_training(trainID_5f, validateID_5f, testID_5f, "scaled")
```

```python
def choose_best_threshold(model, x_val, y_val):
    # Predict Validation Data 
    y_pred_prob = model.predict_proba(x_val)
    
    best_f1score = 0
    best_threshold = 0
    for custom_threshold in np.arange(0.40, 0.60, 0.01):
        custom_threshold = round(custom_threshold, 2)
        y_pred = (y_pred_prob[:,1] >= custom_threshold).astype(int)
        f1 = metrics.f1_score(y_val, y_pred)
        if f1 > best_f1score:
            best_f1score = f1
            best_threshold = custom_threshold
    return best_threshold, best_f1score


def evaluate_model_with_best_threshold(mode, pidList, model, x_test, y_test, best_threshold):
    ### Calculate area under curve (AUC)
    y_pred_prob = model.predict_proba(x_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_prob)
    auc = metrics.roc_auc_score(y_test, y_pred_prob)
    
    ### Predict Test data
    y_pred_prob = model.predict_proba(x_test)
    y_pred = (y_pred_prob[:,1] >= best_threshold).astype(int)
    
    prob_class0 = y_pred_prob[:, 0]
    prob_class1 = y_pred_prob[:, 1]

    ### Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)
    kappa = metrics.cohen_kappa_score(y_test, y_pred)

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)
    
    ### Calculate the corrections
    n_corrections = 0
    for i in range(len(pidList)):
        pid = pidList[i]
        pred = y_pred[i]
        actual = y_test[i]
        
        p1 = prob_class0[i]
        p2 = prob_class1[i]

        if mode=="SVM":
            SVM_pred_result[int(pid)] = {}
            SVM_pred_result[int(pid)]['prob_class0'] = round(p1, 4) ## NC
            SVM_pred_result[int(pid)]['prob_class1'] = round(p2, 4) ## AD
            SVM_pred_result[int(pid)]['prediction'] = pred
            SVM_pred_result[int(pid)]['actual'] = actual
        if pred==actual:
            n_corrections += 1
    return {'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'kappa': kappa, 
            'fpr': fpr, 'tpr': tpr, 'auc': auc, 'cm': cm}, n_corrections


def show_confusion_matrix(mode):
    actual_label_List = []
    predict_label_List = []
    if mode == "SVM":
        for key in SVM_pred_result:
            pred = SVM_pred_result[key]['prediction']
            actual = SVM_pred_result[key]['actual']
            predict_label_List.append(pred)
            actual_label_List.append(actual)
    
    actual_label_array = np.array(actual_label_List)
    predict_label_array = np.array(predict_label_List)

    ### F1-Score
    print(f'==============  Classification Report  ==============')
    print(classification_report(actual_label_array, predict_label_array))

    ### Confusion Matrix
    matrix = confusion_matrix(actual_label_array, predict_label_array)
    TN, FP, FN, TP = matrix.ravel()

    value_accuracy = round((TP+TN)/(TP+TN+FP+FN), 4)
    value_sensitivity = round(TP/(TP+FN), 4)
    value_specificity = round(TN/(TN+FP), 4)
    value_gmeans = round(np.sqrt(value_sensitivity*value_specificity), 4)
    value_precision = round(TP/(TP+FP), 4)
    value_f1_score = round(2*value_precision*value_sensitivity/(value_precision+value_sensitivity), 2)
    value_f2_measure = round(5*value_precision*value_sensitivity / (4*value_precision+value_sensitivity), 2)

    print(f'==================  Performance  ====================')
    print(f'0) TN: {TN}, FP: {FP}, FN: {FN}, TP: {TP}')
    print(f'1) Accuracy: {round(value_accuracy*100, 2)}%')
    print(f'2) Sensitivity (Recall): {round(value_sensitivity*100, 2)}%')
    print(f'3) Specificity: {round(value_specificity*100, 2)}%')
    print(f'4) G-Means: {round(value_gmeans*100, 2)}%')
    print(f'5) Precision: {round(value_precision*100, 2)}%')
    print(f'6) F1-Score: {value_f1_score}')
    print(f'7) F2-Measure: {value_f2_measure}\n')
    
    tmp_eval = {'acc': value_accuracy,
                'prec': value_precision,
                'rec': value_sensitivity,
                'f1': value_f1_score}
    
    
    classList = ['NC', 'AD']
    fig, ax = plot_confusion_matrix(conf_mat=matrix, figsize=(3, 3), colorbar=True, class_names=classList, fontcolor_threshold=0.7)
    plt.xlabel('Predictions', fontsize=10)
    plt.ylabel('Actuals', fontsize=10)
    plt.title('Turning Subtask', fontsize=12)
    plt.show()
    return tmp_eval


def training_model_with_validation(mode, train_5f, val_5f, test_5f):
    ### Choose best "Regularization parameter"
    # param_grid = [0.001, 0.01, 0.1, 1, 10, 100]
    param_grid = [1] 

    for c in param_grid:
        for i in range(n_FOLD):
            print(f'------------------------------------------------------[Fold {i+1}] C={c} -----------------------------------------------------------\n')
            train_df = train_5f[i]
            val_df = val_5f[i]
            test_df = test_5f[i]
            
            ### Copying original dataframe
            train_df_ready = train_df.copy()
            val_df_ready = val_df.copy()
            test_df_ready = test_df.copy()
            
            X_train = train_df_ready.drop(['pid', 'label'], axis=1).values
            y_train = train_df_ready['label'].values
            
            X_val = val_df_ready.drop(['pid', 'label'], axis=1).values
            y_val = val_df_ready['label'].values
            
            X_test = test_df_ready.drop(['pid', 'label'], axis=1).values
            y_test = test_df_ready['label'].values
            
            if mode == "SVM": 
                best_auc = 0
                best_model_idx = 0
                best_threshold = 0
                
                for j in range(n_MODEL):
                    print(f'Training for [Fold {i+1}] model {j+1} ...')
                    
                    # model = svm.SVC(kernel='linear', random_state=42, probability=True) #original
                    # model = svm.SVC(kernel='linear', random_state=seed_value, probability=True, verbose=True) #[RandomSeed]
                    model = svm.SVC(kernel='linear', random_state=seed_value, C=c, probability=True) #[RandomSeed]
                    model.fit(X_train, y_train)
                    
                    ### choose best model performance with validation data
                    PidList_val = validateID_5f[i]
                    yPred_proba_val = model.predict_proba(X_val)[::,1]
                    
                    fpr, tpr, _ = metrics.roc_curve(y_val, yPred_proba_val)
                    aucScore_val = metrics.roc_auc_score(y_val, yPred_proba_val)
                    
                    if aucScore_val>best_auc:
                        best_auc = aucScore_val 
                        best_model_idx = j+1
                        best_model = model
                
                best_threshold, best_f1score = choose_best_threshold(best_model, X_val, y_val)
                print(f'\n -> best_Threshold: {best_threshold}')
                print(f' -> best_F1-Score: {best_f1score}\n')
                best_threshold_5f_list.append(best_threshold)

                ### Save SVM model
                # with open(f'{SAVE_SVM_MODEL}model_{i+1}.pkl', 'wb') as f:
                #     pickle.dump(best_model, f)
                
                ### Plot SVM classification result
                # plot_SVM_classification_figure(i, best_model, X_test, y_test)

                pidList = testID_5f[i]
                # svm_eval, svm_corrections = evaluate_model("SVM", pidList, best_model, X_test, y_test)
                svm_eval, svm_corrections = evaluate_model_with_best_threshold("SVM", pidList, best_model, X_test, y_test, best_threshold)

                key_name = 'Fold_' + str(i+1)
                SVM_metrics_result[key_name] = {}
                SVM_metrics_result[key_name]['Corrections'] = svm_corrections
                SVM_metrics_result[key_name]['Accuracy'] = svm_eval['acc']
                SVM_metrics_result[key_name]['Precision'] = svm_eval['prec']
                SVM_metrics_result[key_name]['Recall'] = svm_eval['rec']
                SVM_metrics_result[key_name]['F1-Score'] = svm_eval['f1']
                SVM_metrics_result[key_name]['AUC'] = svm_eval['auc']
                SVM_metrics_result[key_name]['CM'] = svm_eval['cm']


                ### [update] 20240516: Record the AUC for analysis
                # yPred_proba_test = best_model.predict_proba(X_test)[::,1]
                # fpr, tpr, _ = metrics.roc_curve(y_test, yPred_proba_test)
                # aucScore_test = metrics.roc_auc_score(y_test, yPred_proba_test)
                # aucScore_5fold.append(aucScore_test)



                # ### Model Explanation on Testing data
                # interpretor = lime_tabular.LimeTabularExplainer(training_data = np.array(X_train),  
                #                                                 feature_names = ['period1', 'period2'],  
                #                                                 mode = 'classification',
                #                                                 class_names=['NC', 'AD'])  
                # for test_idx in range(1):
                #     tmp_pid = pidList[test_idx]
                #     tmp_label = Label_dict[int(tmp_pid)]
                #     print(f'pid: {tmp_pid}, label: {tmp_label}')
                #     exp = interpretor.explain_instance(data_row=X_test[test_idx], 
                #                                        predict_fn=best_model.predict_proba)  
                #     print(f' (1) LIME Explanations.')
                #     exp.show_in_notebook(show_table=True)
                #     sp_exp = submodular_pick.SubmodularPick(interpretor, 
                #                                             np.array(X_test),
                #                                             predict_fn=best_model.predict_proba,
                #                                             sample_size=10)
                #     print('\n (2) SP-LIME Explanations.')
                #     [exp.show_in_notebook() for exp in sp_exp.sp_explanations]
                #     # [exp.as_pyplot_figure() for exp in sp_exp.sp_explanations]


                ### [2024/02/24] Visualize decision region on 【Testing Data】 ---------------------------------------------
                ### Reference: https://medium.com/jameslearningnote/%E8%B3%87%E6%96%99%E5%88%86%E6%9E%90-%E6%A9%9F%E5%99%A8%E5%AD%B8%E7%BF%92-%E7%AC%AC3-4%E8%AC%9B-%E6%94%AF%E6%8F%B4%E5%90%91%E9%87%8F%E6%A9%9F-support-vector-machine-%E4%BB%8B%E7%B4%B9-9c6c6925856b
                # visualize_decision_regions(X_test, y_test, classifier=best_model)
                # plt.xlabel('Period1 (time)')
                # plt.ylabel('Period2 (time)')
                # plt.title(f'[Fold {i+1}] Decision Region on Testing Data')
                # plt.legend(loc='upper right')
                # plt.show()
                
                ### [2024/02/24] Visualize hyperplane and support vectors on 【Training Data】 -----------------------------
                ### Reference: https://github.com/christianversloot/machine-learning-articles/blob/main/how-to-visualize-support-vectors-of-your-svm-classifier.md
#                 index_AD_data = np.where(y_train == 1)[0].tolist()
#                 index_NC_data = np.where(y_train == 0)[0].tolist()
#                 X_classAD, y_classAD = X_train[index_AD_data, :], y_train[index_AD_data] # Get AD data
#                 X_classNC, y_classNC = X_train[index_NC_data, :], y_train[index_NC_data] # Get NC data
                
#                 support_vector_indices = best_model.support_       # Get support vector indices
#                 support_vectors_per_class = best_model.n_support_  # Get number of support vectors per class
#                 support_vectors = best_model.support_vectors_      # Get support vectors themselves
                
#                 w = best_model.coef_[0]                            # Get the separating hyperplane
#                 a = -w[0] / w[1]                                   # Get the slope of hyperplane
#                 xx = np.linspace(10, 110)
#                 yy = a * xx - (best_model.intercept_[0]) / w[1]
                
#                 b = support_vectors[0]                             # Get the first support vector which must be POSITIVE (AD)
#                 yy_down = a * xx + (b[1] - a * b[0])               # Get the parallele to the separating hyperplane that pass through the support vector
#                 b = support_vectors[-1]                            # Get the last support vector which must be MEGATIVE (NC)
#                 yy_up = a * xx + (b[1] - a * b[0])

#                 plt.figure(figsize=(5, 3))
#                 plt.scatter(X_classAD[:, 0], X_classAD[:, 1], color='blue', label='AD')
#                 plt.scatter(X_classNC[:, 0], X_classNC[:, 1], color='green', label='NC')
#                 # plt.scatter(support_vectors[:,0], support_vectors[:,1], color='red', label='Support') # Plot support vectors
#                 plt.legend()
#                 plt.plot(xx, yy, linewidth=2, color='black')  # Plot separating hyperplane
#                 plt.plot(xx, yy_down, 'k--')                # Plot margin parallel line
#                 plt.plot(xx, yy_up, 'k--')                  # Plot margin parallel line
                
#                 plt.title(f'[Fold {i+1}] Visualize training data with hyperplane')
#                 plt.xlabel('Period1 (time)')
#                 plt.ylabel('Period2 (time)')
#                 plt.show()
                
                ### [2024/02/24] Visualize feature importance on 【Training Data】 -----------------------------
                ### Reference: https://copyprogramming.com/howto/how-to-get-feature-importance-from-sklearn-svm
                # feature_names = ['Period1', 'Period2']
                # n_class = 2
                # n_feature = 2
                # importance, fnames = zip(*sorted(zip(best_model.coef_[0], feature_names)))
                # print(f'\nImportance: {importance}')
                # print(f'\nfnames: {fnames}')
                # plt.figure(figsize=(5, 3))
                # plt.barh(range(len(fnames)), importance, align='center')
                # plt.yticks(range(len(fnames)), fnames)
                # plt.title(f'[Fold {i+1}] Feature Importaance')
                # plt.show()

                ### [2024/02/27] SHAP -------------------------------------------------------------------------
                ### Reference: https://ithelp.ithome.com.tw/articles/10330115
                SHAP_explainer = shap.KernelExplainer(model=best_model.predict_proba, data=X_train[:50], link='logit')
                shap_values = SHAP_explainer.shap_values(X=X_test)
                x_feature_names = ['period 1', 'period 2']
                y_label_names = ['Non-AD', 'AD']
                shap.summary_plot(shap_values, X_test, plot_type="bar", class_names=y_label_names, feature_names=x_feature_names, plot_size=(6, 2), class_inds='original')

                # print(f'\nHow important the model is in predicting 【NC】')
                # shap.summary_plot(shap_values[0], X_test, feature_names=x_feature_names, plot_size=(6, 2))
                print(f'\nHow important the model is in predicting 【AD】')
                shap.summary_plot(shap_values[1], X_test, feature_names=x_feature_names, plot_size=(6, 2))
                
                ### [2024/02/29] LIME -------------------------------------------------------------------------
                # def model_explanation_return_prob(xTest_1f):
                #     p1,p2 = (np.ones(len(xTest_1f))[0] - best_model.predict(xTest_1f)),  best_model.predict(xTest_1f)
                #     prediction = [[x,y] for x,y in zip(p1,p2)]
                #     return np.array(prediction).reshape(len(xTest_1f),2)
                    
                # feature_name_list = ['period1', 'period2']
                # explainer = lime_tabular.LimeTabularExplainer(training_data=np.array(X_train),  
                #                                               feature_names=feature_name_list,
                #                                               mode='classification',
                #                                               class_names = ['NC', 'AD'])
                # for test_idx in range(len(pidList)):
                #     tmp_pid = pidList[test_idx]
                #     tmp_label = Label_dict[int(tmp_pid)]
                #     tmp_pred = SVM_pred_result[int(tmp_pid)]['prediction']
                #     print(f'pid: {tmp_pid}, Label: {tmp_label}, Pred: {tmp_pred}')

                #     exp = explainer.explain_instance(data_row=X_test[test_idx], predict_fn=model_explanation_return_prob)  
                #     exp.show_in_notebook(show_table=True)
                    
                
                ### [2024/02/24] Baseline Comparison: Random Forest --------------------------------------------
                # acc, prec, rec, f1, auc = baseline_comparison("RF", X_train, y_train, pidList, X_test, y_test)
                # print(f'    Baseline Comparison: Random Forest')
                # print(f'       1) Accuracy: {round(acc, 2)}')
                # print(f'       2) Precision: {round(prec, 2)}')
                # print(f'       3) Recall: {round(rec, 2)}')
                # print(f'       4) F1-Score: {round(f1, 2)}')
                # print(f'       5) AUC: {round(auc, 2)}')
                # print(f'-----------------------------------------------------------------------------------------------------------------------\n')
```

```python
metrics_5fold_dict = {}

def load_model_and_evaluate_five_folds_metrics():
    metrics_5fold_dict = {}
    
    for fold_idx in range(n_FOLD):
        testID = testID_5f[fold_idx]
        yPred, yTrue = [], []
        predict_label_weighted_List, actual_label_weighted_List = [], []
        
        for pid in testID:
            pid = int(pid)
            prob = SVM_pred_result[pid]['prob_class1']
            predict_label = SVM_pred_result[pid]['prediction']
            label = Label_dict[pid]
            yPred.append(prob)
            yTrue.append(label)
            predict_label_weighted_List.append(predict_label)
            actual_label_weighted_List.append(label)
            
        predict_label_weighted_array = np.array(predict_label_weighted_List)
        actual_label_weighted_array = np.array(actual_label_weighted_List)
        matrix = confusion_matrix(actual_label_weighted_array, predict_label_weighted_array)
        TN, FP, FN, TP = matrix.ravel()
        value_accuracy = round((TP+TN)/(TP+TN+FP+FN), 4)
        value_sensitivity = round(TP/(TP+FN), 4)
        value_specificity = round(TN/(TN+FP), 4)
        value_precision = round(TP/(TP+FP), 4)
        value_f1_score = round(2*value_precision*value_sensitivity/(value_precision+value_sensitivity), 2)
        auc = roc_auc_score(yTrue, yPred)
        
        classList = ['Non-AD', 'AD']
        matrix = confusion_matrix(actual_label_weighted_array, predict_label_weighted_array)
        TN, FP, FN, TP = matrix.ravel()
        fig, ax = plot_confusion_matrix(conf_mat=matrix, figsize=(2, 2), colorbar=True, class_names=classList, fontcolor_threshold=0.7)
        plt.xlabel('Predictions', fontsize=10)
        plt.ylabel('Actuals', fontsize=10)
        plt.title(f'Fold {fold_idx+1}', fontsize=12)
        plt.show()
        
        
        metrics_5fold_dict[fold_idx] = {}
        metrics_5fold_dict[fold_idx]['AUC'] = round(auc, 4)
        metrics_5fold_dict[fold_idx]['Accuracy'] = value_accuracy
        metrics_5fold_dict[fold_idx]['Sensitivity'] = value_sensitivity
        metrics_5fold_dict[fold_idx]['Specificity'] = value_specificity
        metrics_5fold_dict[fold_idx]['Precision'] = value_precision
        metrics_5fold_dict[fold_idx]['F1-Score'] = value_f1_score
        
    return metrics_5fold_dict


metrics_5fold_dict = load_model_and_evaluate_five_folds_metrics()
```
